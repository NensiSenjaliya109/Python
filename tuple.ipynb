Remove duplicates from tuple

```python
a = (1, 3, 5)
b = (3, 1, 6)
unique = tuple(set(a) ^ set(b))
print(unique)
common = set(a) & set(b)
print(common)
combine = set(a) | set(b)
print(combine)
```

Convert nested tuple into flat tuple

```python
nested_tuple = ((1, 2), (3, 4, 5))

empty_items=[]

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        empty_items.append(item)

print(empty_items)
```
