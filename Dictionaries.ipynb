```python
student = {
    "name": "Nensi",
    "age": 21

}
print(student["name"])
student.get("name")

student["grade"] = "A"
student.get("grade")
print(student)

student.pop("grade")
print(student)
```

```python
##Dictionary Loops

student = {
    "Name": "Nensi",
    "Age": 21,
}
print(student.keys())
print(student.values())
print(student.items())
```

```python
users = {
    101:{
        "name":"John",
        "email":"john@gmail.com"
    },

    102:{
        "name":"Alex",
        "email":"alex@gmail.com"
    }
}


print(users[101]["email"])
```

```python
scores = {"Alice": 85, "Bob": 92, "Charlie": 78, "Diana": 95}
print(max(scores, key=scores.get))
print(max(scores.values()))
print(max(scores.keys()))
print(max(scores.items()))
```

```python
text = "Apple banana apple cherry banana apple"
word_count = {}
for word in text.split():
    word_count[word] = word_count.get(word, 0) + 1
print(word_count)
```
